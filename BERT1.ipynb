```
vocab_file=tokenization.load_vocab('/content/bert/uncased_L-12_H-768_A-12/vocab.txt')
```

```
word_piece_tokenizer= tokenization.WordpieceTokenizer(vocab_file)
```

```
basic_tokenizer= tokenization.BasicTokenizer()
```

```
basic_tokens=basic_tokenizer.tokenize("Hi, How are you embedding?")
```

```
basic_tokens
```

```
word_piece_tokens=[]
for token in basic_tokens:
  word_piece_tokens.append(word_piece_tokenizer.tokenize(token))
```

```
word_piece_tokens
```

```
word_piece_tokenizer.tokenize('embedding')
```

```
tokens_to_ids=[]
for word in word_piece_tokens:
  tokens_to_ids.append(tokenization.convert_tokens_to_ids(vocab=vocab_file,tokens=word))
```
